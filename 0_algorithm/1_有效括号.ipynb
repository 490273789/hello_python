# 有效括号
给定一个只包括“（’，“），'｛，'｝'，•［，" 的字符串§，判断字符串是否有效。
有效字符串需满足：
1.左括号必须用相同类型的右括号闭合：
2.左括号必须以正确的顺序闭合。
3.每个右括号都有一个对应的相同类型的左括号。

```python

input_str = "(){){[]}"

def valid_bracket(input):
    if not isinstance(len(input), int):
        return False
    map = {
        "[":"]",
        "(":")",
        "{":"}"
    }
    list = []
    for char in input:
        if len(list) > 0 and char == map.get(list[len(list) - 1]):
            list.pop()
        else:
            list.append(char)
    return len(list) == 0
    
valid_bracket(input_str) 
```

```python
input_str = "()()){[]}"
def valid(input):
    if(isinstance(len(input)/2, int)):
        return False
    stack = []
    for char in input:
        match char:
            case "(" | "{" | "[":
                stack.append(char)
            case ")":
                if (not stack) or (stack.pop() != "("):
                    return False
            case "}":
                if (not stack) or (stack.pop() != "{"):
                    return False
            case "]":
                if (not stack) or (stack.pop() != "["):
                    return False
    return True if not stack else False

valid(input_str)
```
