26. 删除有序数组中的重复项
给你一个非严格递增排列的数组 nums ，请你原地删除重复出现的元素，使每个元素 只出现一次 ，返回删除后数组的新长度。元素的相对顺序应该保持一致 。然后返回 nums 中唯一元素的个数。
考虑 nums 的唯一元素的数量为 k。去重后，返回唯一元素的数量 k。
nums 的前 k 个元素应包含 排序后 的唯一数字。下标 k - 1 之后的剩余元素可以忽略。

示例 2：
输入：nums = [0,0,1,1,1,2,2,3,3,4]
输出：5, nums = [0,1,2,3,4,_,_,_,_,_]
解释：函数应该返回新的长度 5 ， 并且原数组 nums 的前五个元素被修改为 0, 1, 2, 3, 4 。不需要考虑数组中超出新长度后面的元素。



```python
from typing import List


class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        map = []
        i = 0
        while i < len(nums):
            if nums[i] in map:
                del nums[i]
            else:
                map.append(nums[i])
                i += 1
        return len(nums)
```

```python
class Solution1:
    def removeDuplicates(self, nums: List[int]) -> int:
        left, r = 0, 1
        length = len(nums)
        while r < length:
            if nums[left] != nums[r]:
                if r - left > 1:
                    nums[left + 1] = nums[r]
                left += 1
            r += 1
        return left + 1
```

```python
class Solution2:
    def removeDuplicates(self, nums: List[int]) -> int:
        left, r = 0, 1
        length = len(nums)
        while r < length:
            if nums[left] != nums[r]:
                left += 1
                # if left < r: # 不加这个判断更快
                nums[left] = nums[r]
            r += 1
        return left + 1
```

```python
class Solution4:
    def removeDuplicates(self, nums: List[int]) -> int:
        left = 0
        for r in range(1, len(nums)):
            if nums[r] != nums[r - 1]:
                left += 1
                nums[left] = nums[r]
        return left + 1
```
