在 Python 中，想要实现迭代器，必须弄清楚两个长得很像的词：

1. 可迭代对象 (Iterable):

    - 定义：表面意思是“可以被迭代的东西”。比如 list, str, tuple, dict。
    - 特征：它们内部都有一个 __iter__() 方法。
    - 地位：它们是原材料，是还没有上流水线的菜。

2. 迭代器 (Iterator):

    - 定义：这就是那个“流水线”本身。
    - 特征：
        1. 有 __iter__() 方法（返回它自己）。
        2. 最关键的：有 __next__() 方法。
        3. 地位：它是真正的执行者。当你调用 next() 时，它给你返回下一个数据。

迭代list
1. 迭代器是一次性的消耗品，不能回头，用过一次就不能使用了


```python
my_list = [1,2,3,4]

my_iterator = iter(my_list)

print("迭代器准备就绪。。。。")
while True:
    text = input()
    if text == "":
        try:
            res = next(my_iterator)
            print(res)
        except StopIteration as e:
            print(f"{e}迭代结束。。。")
            break
    else:
        break
    

```
